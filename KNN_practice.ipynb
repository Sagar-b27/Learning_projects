```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model,preprocessing

data = pd.read_csv("car.data")
print(data.head())

# Processing non integer data

ppd = preprocessing.LabelEncoder()

buying = ppd.fit_transform(list(data["buying"]))
maint = ppd.fit_transform(list(data["maint"]))
door = ppd.fit_transform(list(data["doors"]))
persons = ppd.fit_transform(list(data["persons"]))
lug_boot = ppd.fit_transform(list(data["lug_boot"]))
safety = ppd.fit_transform(list(data["safety"]))
cl = ppd.fit_transform(list(data["class"]))


predict = "class"

# Converting numpy array into list for processing data

X = list(zip(buying, maint, door, persons, lug_boot, safety))
Y = list(cl)

X_train,X_test,Y_train,Y_test = sklearn.model_selection.train_test_split(X,Y,test_size=0.1)

model = KNeighborsClassifier(n_neighbors=7)

model.fit(X_train,Y_train)
accuracy = model.score(X_test,Y_test)
print(accuracy)

predicted = model.predict(X_test)
classes = ["unacc","acc","good","vgood"]

for i in range(len(X_test)):
  print("Predicted: ", classes[predicted[i]], "Data: ", X_test[i], "Actual_value: ", classes[Y_test[i]])
  # return distances of nearest neighbors and the index 
  n = model.kneighbors([X_test[i]], 7)
  print("Distances from nearest neighbors : ", n[0])
  print("Position of nearest neighbors : ", n[1])
  print("---------------------------------------")
```
