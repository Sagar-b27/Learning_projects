```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

data = keras.datasets.fashion_mnist

#Sepaarting data into training and testing data
(train_images, train_labels), (test_images, test_labels) = data.load_data()

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

#Shrinking gray scale values between 0 and 1
train_images = train_images/255.0
test_images = test_images/255.0

#Constructing neural network
#Flatten will merge all the arrays into a single array which is passed as input
model = keras.Sequential([
      keras.layers.Flatten(input_shape = (28,28)),
      keras.layers.Dense(128, activation="relu"),
      keras.layers.Dense(10, activation="softmax")   
      ])

model.compile(optimizer="adam", loss="SparseCategoricalCrossentropy", metrics=["accuracy"])

#Epochs is the number of times a same data input is shown to the network
model.fit(train_images, train_labels, epochs = 5)

#test_loss, test_accuracy = model.evaluate(test_images, test_labels)
#print(test_accuracy)

prediction = model.predict(test_images)

for i in range(10):
  plt.grid(False)
  plt.imshow(test_images[i], cmap = plt.cm.binary)
  plt.xlabel("Actual: " + class_names[test_labels[i]])
  plt.title("Prediction: " + class_names[np.argmax(prediction[i])])
  plt.show()

```
